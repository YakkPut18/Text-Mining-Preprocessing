# Preprocessing

Removing punctuations

```python
import string

def remove_punctuation(text):
  """Menghapus semua tanda baca dari sebuah string."""
  if isinstance(text, str):
    return text.translate(str.maketrans('', '', string.punctuation))
  return text

# Asumsikan kolom teks Anda bernama 'Tweet'
# Ganti 'Tweet' dengan nama kolom teks Anda yang sebenarnya jika berbeda
df['cleaned_text_column'] = df['Tweet'].apply(remove_punctuation)

# Tampilkan beberapa baris pertama dengan kolom yang telah dibersihkan
display(df[['Tweet', 'cleaned_text_column']].head())
```

Removing URLs

```python
import re

def remove_urls(text):
  """Menghapus URL dari sebuah string."""
  if isinstance(text, str):
    # Regular expression to find URLs
    url_pattern = re.compile(r'https?://\S+|www\.\S+')
    return url_pattern.sub(r'', text)
  return text

# Terapkan fungsi ke kolom 'cleaned_text_column'
df['text_without_urls'] = df['cleaned_text_column'].apply(remove_urls)

# Tampilkan beberapa baris pertama dengan kolom baru
display(df[['cleaned_text_column', 'text_without_urls']].head())
```

Removing stop words

```python
import nltk
from nltk.corpus import stopwords

# Unduh daftar stop words Bahasa Indonesia
nltk.download('stopwords')

# Dapatkan stop words Bahasa Indonesia
stop_words_id = set(stopwords.words('indonesian'))

def remove_stopwords(text):
  """Menghapus stop words Bahasa Indonesia dari sebuah string."""
  if isinstance(text, str):
    words = text.split()  # Tokenisasi sederhana dengan memisahkan berdasarkan spasi
    filtered_words = [word for word in words if word.lower() not in stop_words_id]
    return ' '.join(filtered_words)
  return text

# Terapkan fungsi ke kolom 'text_without_urls'
df['text_without_stopwords'] = df['text_without_urls'].apply(remove_stopwords)

# Tampilkan beberapa baris pertama dengan kolom baru
display(df[['text_without_urls', 'text_without_stopwords']].head())
```

Removing mention & hastag

```python
import re

def remove_mentions_hashtags(text):
  """Removes mentions and hashtags from a string."""
  if isinstance(text, str):
    # Regular expression to find mentions (@username) and hashtags (#hashtag)
    pattern = re.compile(r'@\S+|#\S+')
    return pattern.sub(r'', text)
  return text

# Apply the function to the 'text_without_stopwords' column
df['text_without_mentions_hashtags'] = df['text_without_stopwords'].apply(remove_mentions_hashtags)

# Tampilkan beberapa baris pertama dengan kolom baru
display(df[['text_without_stopwords', 'text_without_mentions_hashtags']].head())
```

Removing Space

```python
import re

def remove_extra_spaces(text):
  """Removes extra spaces from a string."""
  if isinstance(text, str):
    # Ubah menjadi spasi tunggak
    return re.sub(r'\s+', ' ', text).strip()
  return text

# Terapkan ke fungsi 'text_without_mentions_hashtags'
df['text_without_extra_spaces'] = df['text_without_mentions_hashtags'].apply(remove_extra_spaces)

# Tampilkan beberapa baris pertama dengan kolom baru
display(df[['text_without_mentions_hashtags', 'text_without_extra_spaces']].head())
```

Removing duplicate letter

```python
import re

def remove_repeated_chars(text):
    """Menghapus karakter berulang (misalnya, 'aaaa' menjadi 'aa')."""
    if isinstance(text, str):
        # Regex ini mencari karakter apa pun (\w) diikuti oleh dua atau lebih karakter yang sama (\1{2,}).
        # Ini menggantinya dengan karakter diikuti oleh satu instance dari dirinya sendiri (\1\1).
        # Ini mempertahankan variasi seperti "aa" tetapi mengurangi "aaa", "aaaa", dll.
        # Untuk kasus seperti "nggakkkk", mungkin diperlukan regex yang lebih spesifik atau beberapa kali pass.
        # Pendekatan yang lebih kuat mungkin melibatkan pemeriksaan terhadap kamus, tetapi ini adalah metode regex umum.
        pattern = re.compile(r'(\w)\1{2,}')
        return pattern.sub(r'\1\1', text)
    return text

# Terapkan fungsi ke kolom 'text_without_extra_spaces'
# Atau terapkan ke kolom sebelum stemming/lemmatization tergantung urutan yang diinginkan
df['text_without_repeated_chars'] = df['text_without_extra_spaces'].apply(remove_repeated_chars)

# Tampilkan beberapa baris pertama dengan kolom baru
display(df[['text_without_extra_spaces', 'text_without_repeated_chars']].head())
```

```python
import re

def remove_extra_spaces(text):
  """Menghapus spasi berlebih dari sebuah string."""
  if isinstance(text, str):
    # Mengganti beberapa spasi dengan satu spasi dan menghapus spasi di awal/akhir
    return re.sub(r'\s+', ' ', text).strip()
  return text

# Terapkan fungsi ke kolom 'text_without_mentions_hashtags'
df['text_without_extra_spaces'] = df['text_without_mentions_hashtags'].apply(remove_extra_spaces)

# Tampilkan beberapa baris pertama dengan kolom baru
display(df[['text_without_mentions_hashtags', 'text_without_extra_spaces']].head())
```

Lowercasing

```python
def lowercase_text(text):
  """Mengubah string menjadi huruf kecil."""
  if isinstance(text, str):
    return text.lower()
  return text

# Terapkan fungsi ke kolom 'text_without_stopwords'
df['text_lowercase'] = df['text_without_stopwords'].apply(lowercase_text)

# Tampilkan beberapa baris pertama dengan kolom baru
display(df[['text_without_stopwords', 'text_lowercase']].head())
```

Tokenization

```python
import nltk
from nltk.tokenize import word_tokenize

# Unduh resource yang diperlukan untuk tokenisasi
nltk.download('punkt')
nltk.download('punkt_tab') # Unduh punkt_tab secara eksplisit

def tokenize_text(text):
  """Melakukan tokenisasi pada sebuah string."""
  if isinstance(text, str):
    return word_tokenize(text)
  return text

# Terapkan fungsi ke kolom 'text_lowercase'
df['text_tokenized'] = df['text_lowercase'].apply(tokenize_text)

# Tampilkan beberapa baris pertama dengan kolom baru
display(df[['text_lowercase', 'text_tokenized']].head())
```

Stemming

```python
import re
import nltk
from nltk.tokenize import word_tokenize
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory

# Pastikan sudah download resource
nltk.download('punkt')

# Inisialisasi stemmer bahasa Indonesia
factory = StemmerFactory()
stemmer = factory.create_stemmer()

# Contoh: gunakan dataset bernama data
# data = pd.read_csv('INA_TweetsPPKM_Labeled_Pure.csv', ...)

# Bersihkan teks dasar
def clean_text(text):
    text = str(text).lower()
    text = re.sub(r"http\S+|www\S+", "", text)  # hapus URL
    text = re.sub(r"@\w+|#\w+", "", text)       # hapus mention & hashtag
    text = re.sub(r"[^a-z\s]", "", text)        # hapus tanda baca & angka
    text = re.sub(r"\s+", " ", text).strip()    # hapus spasi berlebih
    return text

# Tokenization
df['clean_tweet'] = df['Tweet'].apply(clean_text)
df['text_tokenized'] = df['clean_tweet'].apply(word_tokenize)

# Stemming function
def stemming(tokens):
    return [stemmer.stem(word) for word in tokens]

# Apply stemming
df['text_stemmed'] = df['text_tokenized'].apply(stemming)

# Lihat hasil
df[['Tweet', 'clean_tweet', 'text_tokenized', 'text_stemmed']].head()
```

Lemman

```python
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory
from nltk.stem import WordNetLemmatizer
from langdetect import detect
import nltk
import re

nltk.download('wordnet')
nltk.download('omw-1.4')

# Inisialisasi alat
factory = StemmerFactory()
indo_stemmer = factory.create_stemmer()
eng_lemmatizer = WordNetLemmatizer()

def mixed_lemmatize(text):
    words = re.findall(r'\b\w+\b', text.lower())
    result = []
    for word in words:
        try:
            lang = detect(word)
        except:
            lang = "id"  # default jika gagal deteksi

        if lang == "id":
            lemma = indo_stemmer.stem(word)
        else:
            lemma = eng_lemmatizer.lemmatize(word)
        result.append(lemma)
    return ' '.join(result)

# Contoh
text = "PPKM is extended again guys, please tetap di rumah ya"
print(mixed_lemmatize(text))
```

```python
!pip install gensim

from gensim.models import Word2Vec
import warnings
warnings.filterwarnings('ignore')

tokenized_news = df['text_stemmed'].tolist()

# Melatih model Word2Vec
# Parameters:
# - sentences: data teks yang sudah ditokenisasi.
# - vector_size: dimensi dari vektor kata yang akan dihasilkan (misal: 100).
# - window: jarak maksimum antara kata saat ini dan kata yang diprediksi dalam satu kalimat.
# - min_count: mengabaikan semua kata dengan frekuensi total lebih rendah dari ini.
# - workers: jumlah thread worker untuk melatih model.

word2vec_model = Word2Vec(sentences=tokenized_news, vector_size=100, window=5, min_count=1, workers=4)

print("Word2Vec model trained successfully!")
```

```python
# Menampilkan informasi tentang model Word2Vec yang telah dilatih
print("Objek Word2Vec KeyedVectors (word2vec_model.wv):")
print(word2vec_model.wv)
vocab_size = len(word2vec_model.wv.key_to_index)
vector_dimension = word2vec_model.wv.vector_size
print(f"Bentuk Word2Vec (kosakata, dimensi vektor): ({vocab_size}, {vector_dimension})")

# Tampilkan beberapa contoh saja jika kosakatanya sangat besar
print("Nama fitur (kosakata Word2Vec):")
feature_names_w2v = list(word2vec_model.wv.key_to_index.keys())
print(feature_names_w2v[:50]) # Menampilkan 50 kata pertama
if len(feature_names_w2v) > 50:
    print(f"... dan {len(feature_names_w2v) - 50} kata lainnya.")
```
